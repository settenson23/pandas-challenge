```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(csv_path)
purchase_data_df.head()
```

```python
#Display the total number of players
Player_Count_df = len(purchase_data_df["SN"].unique())
Player_Count_df
```

```python
#Number of Unique Items
Unique_Items = len(purchase_data_df["Item Name"].unique())

#Average Purchase Price
Average_Price = purchase_data_df["Price"].mean()

#Total Number of Purchases
Total_Purchases = len(purchase_data_df["Item Name"])

#Total Revenue
Total_Revenue = purchase_data_df["Price"].sum()

#Summary Table
Summary_Purchase_Analysis_df = pd.DataFrame({"Number of Unique Items": [Unique_Items], "Average Purchase Price": [Average_Price], "Total Number of Purchases": [Total_Purchases], "Total Revenue": [Total_Revenue]})

#Formatting of Summary Table
Summary_Purchase_Analysis_df["Average Purchase Price"] = Summary_Purchase_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Summary_Purchase_Analysis_df["Total Revenue"] = Summary_Purchase_Analysis_df["Total Revenue"].map("${:.2f}".format)
Summary_Purchase_Analysis_df
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed
gender_count_df = purchase_data_df.groupby("Gender")
each_gender_count_df = gender_count_df["SN"].nunique() 
gender_percentages_df = (each_gender_count_df / Player_Count_df) * 100

#Summary Table
Gender_Analysis_df = pd.DataFrame({"Total Count": each_gender_count_df, "Percentage of Players": gender_percentages_df})

#Sort the total purchase value column in descending order
Gender_Analysis_df = Gender_Analysis_df.sort_values(["Total Count"], ascending=False)

#Formatting of Summary Table
Gender_Analysis_df ["Percentage of Players"] = Gender_Analysis_df["Percentage of Players"].map("{:.2f}%".format)
Gender_Analysis_df
```

```python
#Dataframe sorted by gender
gender_df = purchase_data_df.groupby(["Gender"])

#Purchase count by gender
purchase_count_df = gender_df["SN"].count()

#Avg. purchase price by gender
avg_purchase_price_df = gender_df["Price"].mean()

#Total Purchase Value by gender
purchase_total_df = gender_df["Price"].sum()

# Average purchase total by gender divivded by purchase count by unique shoppers
avg_purchase_per_person_of_gender_df = (purchase_total_df / each_gender_count_df)

#Summary Table
Gender_Purchase_Analysis_df = pd.DataFrame({"Purchase Count": purchase_count_df, "Average Purchase Price": avg_purchase_price_df, "Total Purchase Value": purchase_total_df, "Average Purchase Total Per Person": avg_purchase_per_person_of_gender_df})

#Sort the total purchase value column in descending order
Gender_Purchase_Analysis_df = Gender_Purchase_Analysis_df.sort_values(["Purchase Count"], ascending=False)

#Formatting
Gender_Purchase_Analysis_df["Average Purchase Price"] = Gender_Purchase_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Gender_Purchase_Analysis_df["Total Purchase Value"] = Gender_Purchase_Analysis_df["Total Purchase Value"].map("${:.2f}".format)
Gender_Purchase_Analysis_df["Average Purchase Total Per Person"] = Gender_Purchase_Analysis_df["Average Purchase Total Per Person"].map("${:.2f}".format)
Gender_Purchase_Analysis_df
```

```python
#Establish bins for ages
#max_age_df = purchase_data_df["Age"].max
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 46]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins
purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], bins, labels=bin_names, include_lowest=True)
age_groups_df = purchase_data_df.groupby("Age Groups")

#Calculate the total by age group
total_by_age_group_df = age_groups_df["SN"].nunique()

#Caluclate percentage by age group
pecentage_by_age_group_df = (total_by_age_group_df/Player_Count_df) * 100

#Summary Table
Age_Demographics_df = pd.DataFrame({"Total Count": total_by_age_group_df, "Percentage of Players": pecentage_by_age_group_df})

#Formatting
Age_Demographics_df["Percentage of Players"] = Age_Demographics_df["Percentage of Players"].map("{:.2f}%".format)
Age_Demographics_df
```

```python
#Dataframe sorted by age group
age_groups_df = purchase_data_df.groupby(["Age Groups"])

#Purchase count by age group
purchase_count_df = age_groups_df["SN"].count()

#Avg. purchase price by age group
avg_purchase_price_df = age_groups_df["Price"].mean()

#Total purchase value by age group
purchase_total_df = age_groups_df["Price"].sum()

#Average purchase per person per age group
avg_purchase_per_person_of_age_group_df = (purchase_total_df / total_by_age_group_df)

#Summary Table
Age_Purchase_Analysis_df = pd.DataFrame({"Purchase Count": purchase_count_df, "Average Purchase Price": avg_purchase_price_df, "Total Purchase Value": purchase_total_df, "Average Purchase Total Per Person": avg_purchase_per_person_of_age_group_df})

#Formatting
Age_Purchase_Analysis_df["Average Purchase Price"] = Age_Purchase_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Age_Purchase_Analysis_df["Total Purchase Value"] = Age_Purchase_Analysis_df["Total Purchase Value"].map("${:.2f}".format)
Age_Purchase_Analysis_df["Average Purchase Total Per Person"] = Age_Purchase_Analysis_df["Average Purchase Total Per Person"].map("${:.2f}".format)
Age_Purchase_Analysis_df
```

```python
#Sort dataframe by each player
player_df = purchase_data_df.groupby("SN")

#Purchase Count of each player
each_player_purchase_count_df = player_df["Purchase ID"].count()

#Average Purchase Price of each player
each_player_avg_purchase_price_df = player_df["Price"].mean()

#Total Purchase Value of each players
each_player_total_purchase_value_df = player_df["Price"].sum()

#Summary Table
Top_Spenders_df = pd.DataFrame({"Purchase Count": each_player_purchase_count_df, "Average Purchase Price": each_player_avg_purchase_price_df, "Total Purchase Value": each_player_total_purchase_value_df})

#Sort the total purchase value column in descending order
Top_Spenders_df = Top_Spenders_df.sort_values(["Total Purchase Value"], ascending=False)

#Formatting
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders_df["Total Purchase Value"] = Top_Spenders_df["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders_df.head()
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
item_info_df = purchase_data_df[["Item ID", "Item Name", "Price"]]

#Group by Item ID and Item Name
item_df = item_info_df.groupby(["Item ID", "Item Name"])

#Purchase Count of Each Item
item_purchase_count_df = item_df["Price"].count()

#Total Purchase Value
item_total_purchase_value_df = item_df["Price"].sum()

#Item Price
item_price_df = (item_total_purchase_value_df / item_purchase_count_df)

#Summary Table
Most_Popular_Items_df = pd.DataFrame({"Purchase Count": item_purchase_count_df, "Item Price": item_price_df, "Total Purchase Value": item_total_purchase_value_df})

#Sort the purchase count column in descending order
Most_Popular_Items_df = Most_Popular_Items_df.sort_values(["Purchase Count"], ascending=False)

#Formatting
Most_Popular_Items_df["Item Price"] = Most_Popular_Items_df["Item Price"].map("${:.2f}".format)
Most_Popular_Items_df["Total Purchase Value"] = Most_Popular_Items_df["Total Purchase Value"].map("${:.2f}".format)
Most_Popular_Items_df.head()
```

```python
#Summary Table
Most_Popular_Items_df = pd.DataFrame({"Purchase Count": item_purchase_count_df, "Item Price": item_price_df, "Total Purchase Value": item_total_purchase_value_df})

#Sort the total purchase value column in descending order
Most_Popular_Items_df = Most_Popular_Items_df.sort_values(["Total Purchase Value"], ascending=False)

#Formatting
Most_Popular_Items_df["Item Price"] = Most_Popular_Items_df["Item Price"].map("${:.2f}".format)
Most_Popular_Items_df["Total Purchase Value"] = Most_Popular_Items_df["Total Purchase Value"].map("${:.2f}".format)
Most_Popular_Items_df.head()
```
